```python
#Manipulación de datos, generación y modificación de dataframes
import pandas as pd
# Importa MinMaxScaler de sklearn para escalar características
from sklearn.preprocessing import MinMaxScaler
```

```python
# Define el nombre del archivo que se va a leer.
lista_archivos=[]
lista_nombres=[]
for i in range (0,10,1):
  lista_nombres.append('HuGaDB_v1_bicycling_01_0' + str(i)+'.txt')

for i in range(10,19,1):
  lista_nombres.append('HuGaDB_v1_bicycling_01_' + str(i)+'.txt')

print(lista_nombres)
```

```python
archivo_salida = 'Ciclismo'
lista_dataframes = []

for i in range(0, 10,1):
    # Leer todas las líneas del archivo de entrada
    with open(lista_nombres[i], 'r') as f:
        lineas = f.readlines()

    # Excluir las tres primeras líneas
    lineas_sin_tres_primeras = lineas[3:]

    # Escribir las líneas restantes en el archivo de salida
    with open(archivo_salida, 'w') as f:
        f.writelines(lineas_sin_tres_primeras)

    # Leer el contenido del archivo de salida y convertirlo a un DataFrame
    df = pd.read_csv(archivo_salida, sep='\t', header=0)
    lista_dataframes.append(df)  # Agregar el DataFrame a la lista

# Combinar todos los DataFrames en uno solo
df_combinado = pd.concat(lista_dataframes, ignore_index=True)

# Mostrar el DataFrame combinado
df_combinado
```

```python
# Define el tamaño del conjunto de datos de prueba
test_data_size = 1173

# Asigna los valores del DataFrame, excepto los últimos 20, a train_data
train_data = df_combinado[:-test_data_size].values

# Asigna los valores de los últimos 20 registros del DataFrame a test_data
test_data = df_combinado[-test_data_size:].values

print("Train data: \n")
print(train_data)
print('\nTest data: \n')
print(test_data)
```

```python
# Crea una instancia del MinMaxScaler
scaler = MinMaxScaler()

# Ajusta el scaler a los datos de entrenamiento
scaler = scaler.fit(train_data)

# Transforma los datos de entrenamiento utilizando el scaler ajustado
train_data = scaler.transform(train_data)

# Transforma los datos de prueba utilizando el mismo scaler ajustado
test_data = scaler.transform(test_data)

print("Train data: \n")
print(train_data)
print('\nTest data: \n')
print(test_data)
```
